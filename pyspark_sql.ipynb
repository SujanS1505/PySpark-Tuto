```python
from pyspark.sql import SparkSession

spark=SparkSession.builder.appName('func').getOrCreate()

df=spark.read.csv("D:\Study\Centillion\Pyspark\Datasets\household_power_consumption_household_power_consumption.csv",header=True,inferSchema=True)


df.show()
```

```python
from pyspark.sql.types import *
from pyspark.sql.functions import *


date_plus=date_add(df['Time'],5)
date_plus=date_plus.cast(StringType())

concat_col=concat(date_plus,lit('_updated'))

df.select(df['Time'],concat_col.alias('Time_updated')).show()
```

```python
df.sort("Date").show(5)
```

```python
df.sort(df["Date"].desc()).show(5)
 
```

Using SQL Expressions

```python
df.createOrReplaceTempView("sql")
spark.sql("SELECT time AS Original_Date,date_add(Time,5) AS Updated_date FROM sql").show()
```

```python
spark.sql("SELECT Global_intensity from sql WHERE Global_intensity>10.5").show(10)
```
