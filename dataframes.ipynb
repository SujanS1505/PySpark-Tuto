```python
from pyspark.sql import SparkSession

spark=SparkSession.builder.appName("first").getOrCreate()

df=spark.read.csv("D:\Study\Centillion\Pyspark\Datasets\Piece_Dimension.csv",header=True,inferSchema=True)

df.show()
```

```python
df.printSchema()
```

```python
df.columns
```

```python
#single column
print(df.select("Height"))

#multiple columns
print(df.select(["Height","Width","Length"]))
```

```python
print(df.dtypes)

df.describe().show()
```

```python
#Adding a column in a dataframe
df=df.withColumn("Total Dimension",df.Height+df.Width+df.Length)
df.show()
```

```python
#Dropping a column
df=df.drop("Operator")
df.show()
```

```python
df=df.withColumnRenamed("Total Dimension","Sum Dimension")
df.show()
```

```python
#drops rows with null values
df=df.na.drop()  

#drops rows with any null values
df.na.drop(how='any')  

#drops rows with all null values
df.na.drop(how='all')

#drops rows with less than 2 non-null values
df.na.drop(how="any", thresh=2)  

#drops rows with null values in specific columns
df.na.drop(how="any", subset=["Height"])

#fills all null values with "No Value"
df.na.fill("No Value").show() 
```

```python
from pyspark.ml.feature import Imputer

#fills null values with mean value of the column
imputer=Imputer(inputCols=["Height","Width","Length"],outputCols=["Height","Width","Length"]).setStrategy("mean")
imputer.fit(df).transform(df).show()    
```

```python
df.filter("Height > 22").show()
```

```python
df.filter("height>22").select(["Width","height"]).show()
```

```python
df.filter((df["height"] > 22) & (df["Width"] > 50)).show()
```

```python
#inverse filter
df.filter(~(df["height"]>18.5)).show()
```

```python
#groupby
df.groupBy("Height").count().show(10)
```

```python
df.groupBy("Height").max().show(10)
```

```python
df.groupBy("Height").mean().show(10)
```

```python
#aggregate functions
print(df.agg({'Height':'Sum'}).show())
print(df.agg({'Width':'Mean'}).show())
print(df.agg({'Item_No':'count'}).show())
```
